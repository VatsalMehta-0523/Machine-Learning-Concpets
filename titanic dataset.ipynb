## Classification Metrics – One-Line Interpretations

**Accuracy**
- Measures overall correctness of the model; suitable when classes are reasonably balanced and all errors have similar cost.

**Precision**
- Measures how trustworthy positive predictions are; used when false positives are costly and wrong alarms must be minimized.

**Recall**
- Measures how well the model captures all actual positive cases; used when false negatives are costly and missing positives is unacceptable.

**F1-Score**
- Harmonic mean of precision and recall; used when both false positives and false negatives matter and class imbalance exists.

# transformer
what we will do is that we will create differnt transformer for various steps like 
- imputing missing values for age and embarked
- ohe for sex and embarked
- scaling so that every column gets to a specific range
- feature selection
- implement DT

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
#for imputing missing values
trf1 = ColumnTransformer([
        ("impute_age" , SimpleImputer(strategy = "mean") , [2] ) ,
        ("impute_embarked" , SimpleImputer(strategy = "most_frequent" ) , [6])
    ],
    remainder = "passthrough"
)
```

```python
#for one hot encoding of sex and embarked
trf2 = ColumnTransformer(
    [
        ("ohe_embarked_sex" , OneHotEncoder(sparse_output = False , handle_unknown = "ignore") , [2 , 6])
    ]
)
```

```python
#for scaling 
trf3 = ColumnTransformer(
    [
        ("scaling" , MinMaxScaler() ,[0, 2, 3, 4, 5])
    ]
)
```

```python
#for feature selection however not required here but for learning purpose only
from sklearn.feature_selection import SelectKBest , chi2
trf4 = SelectKBest(score_func = chi2 , k = 5 )
```

```python
# for model implementations
from sklearn.tree import DecisionTreeClassifier
trf5 = DecisionTreeClassifier()
```

# pipeline
we will create a pipelines that connects this chains or transformers

```python
from sklearn.pipeline import Pipeline , make_pipeline
pipe = Pipeline ([
    ("trf1" , trf1),
    ("trf2" , trf2),
    ("trf3" , trf3),
    ("trf4" , trf4),
    ("trf5" , trf5) ]
)
```

Pipeline vs make_pipeline (scikit-learn)

Pipeline:
- You manually assign step names
- More control over naming

Syntax:
Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])

make_pipeline:
- Step names are auto-generated
- Less verbose, quicker to write

Syntax:
make_pipeline(
    StandardScaler(),
    LogisticRegression()
)


# why pipe.fit() and not pipe.fit_transform()
so basically if you are using a training a model as transformer step use fit so you can predict 

notes for reference 
- fit() is required to train the final estimator (model)
- Models do not have a transform() method
- fit_transform() works only when the last step is a transformer
- fit() enables predict(), score(), etc.

Rule:
If the pipeline ends with a model → use fit()
If the pipeline ends with a transformer → use fit_transform()

# Machine Learning Workflow / Lifecycle (Reference Guide)

> **Purpose:**  
> This document acts as a reusable mental + practical framework for approaching any machine learning problem.
> The goal is not just to train a model, but to make *correct, explainable, and justified decisions* at every step.

---

## 1. Problem Framing

### 1.1 Define the Task
- What kind of ML problem is this?
  - Classification / Regression / Clustering / etc.
- If classification:
  - Binary or multiclass?
- What exactly should the model predict?

> Always state the task in **one clear sentence**.

---

### 1.2 Decision Context
- How will the model output be used?
  - Human decision support?
  - Fully automated decision?
- Is interpretability important?
- Is this a low-risk or high-risk decision?

These answers influence:
- Model choice
- Evaluation metric
- Complexity tolerance

---

## 2. Data Understanding

### 2.1 Row-Level Meaning
- What does **one row** represent?
  - One person?
  - One flower?
  - One transaction?

If this is unclear, stop and clarify before proceeding.

---

### 2.2 Column-Level Meaning
For each feature:
- What does it represent?
- Numeric or categorical?
- Measured or derived?
- Continuous or discrete?

⚠️ Never treat a column as “just a number” without understanding its meaning.

---

## 3. Exploratory Data Analysis (EDA)

EDA is about **validating assumptions**, not making plots for aesthetics.

---

### 3.1 Structural Checks
- Dataset shape
- Data types
- Missing values
- Duplicate rows

Key question:
> Is this dataset clean by design, or messy by nature?

---

### 3.2 Target Variable Analysis (Do This Early)
- Class distribution (for classification)
- Imbalance issues
- Rare classes

This step directly affects:
- Metric selection
- Sampling strategy
- Baseline expectations

---

### 3.3 Feature Analysis
For each feature:
- Distribution shape (normal, skewed, bounded)
- Scale range
- Presence of outliers

For feature–target relationship:
- Are classes separable?
- Linear or nonlinear patterns?
- Strong overlap or clear boundaries?

This is where **model intuition starts forming**.

---

## 4. Data Preprocessing (Decision-Driven)

Every preprocessing step must answer **WHY**, not just HOW.

---

### 4.1 Missing Values
Questions to ask:
- Why are values missing?
- Random or systematic?
- Should missingness itself be a signal?

Imputation strategy depends on:
- Feature meaning
- Distribution
- Model sensitivity

---

### 4.2 Scaling
Ask before scaling:
- Does the model rely on distances or gradients?
- Are feature scales very different?

General intuition:
- Distance / gradient-based models → scaling matters
- Tree-based models → scaling usually does not matter

---

### 4.3 Encoding
- Nominal vs ordinal categories
- Does numeric order have semantic meaning?

Avoid blind one-hot encoding.

---

## 5. Train–Test Strategy

Decide before modeling:
- Train–test split or cross-validation?
- Stratified split needed?
- Any data leakage risk?

Splitting strategy must reflect:
- Dataset size
- Target distribution
- Real-world usage scenario

---

## 6. Baseline Model (Mandatory Step)

Purpose of a baseline:
- Establish a performance floor
- Detect leakage early
- Build intuition

Examples:
- Logistic Regression
- k-NN
- Naive Bayes
- Simple Decision Tree

If a complex model barely beats the baseline → investigate the data, not the model.

---

## 7. Model Selection

Choose models based on:
- Dataset size
- Feature type
- Linearity vs non-linearity
- Interpretability requirements

You should be able to justify:
> “I chose this model because …”

---

## 8. Evaluation Metrics

Accuracy is a **choice**, not a default.

Ask:
- Cost of false positives vs false negatives?
- Class imbalance?
- Probabilities vs hard labels needed?

Metric must align with **problem cost**, not convenience.

---

## 9. Error Analysis

After evaluation:
- Which samples are misclassified?
- Any pattern in errors?
- Certain classes confused more than others?

This step guides:
- Feature engineering
- Model improvement
- Next experiment direction

---

## 10. Iteration & Insights

Always conclude with:
- What worked?
- What failed?
- What would I try next?

Even small datasets should produce insights.

---

> **Reminder:**  
> Machine Learning is not about finding the “best model”.
> It is about making **correct, explainable, and justified decisions** at every step.
